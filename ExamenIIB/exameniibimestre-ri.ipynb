Ahora viene lo bueno, la información está en la columna "abstract" así que limpiaremos el texto.

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
def limpiar_texto(text):
    text = text.lower()
    
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'[^a-z0-9\s]', '', text)
    
    text = re.sub(r'\s+', ' ', text).strip()
    
    words = text.split()
    words = [w for w in words if w not in stop_words]
    
    return " ".join(words)
```

```python
df_reducido['abstract_clean'] = df_reducido['abstract'].apply(limpiar_texto)
```

```python
print(df_reducido[['abstract', 'abstract_clean']].head())
```

# 2. Representación mediante embeddings

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# 1. Cargar el modelo SBERT
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = df_reducido['abstract_clean'].tolist()

embeddings = model.encode(sentences, batch_size=64, show_progress_bar=True)
```

# 3. Recuperación inicial

```python
!pip install faiss-cpu
```

```python
import faiss

dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)

embeddings_faiss = np.array(embeddings).astype('float32')

# Agregar vectores al índice
index.add(embeddings_faiss)

print(f"Índice FAISS listo con {index.ntotal} vectores.")
```

```python
def buscar_similares(query, k=5):
    query_vector = model.encode([query]).astype('float32')
    
    distancias, indices = index.search(query_vector, k)
    
    return df_reducido.iloc[indices[0]]
```

Recuperamos información con una query.

```python
resultados = buscar_similares("Advancements in large language models and transformers", k=3)
print(resultados[['title', 'abstract']])
```

# 4. Re-ranking de resultados

```python
from sentence_transformers import CrossEncoder

reranker_model = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')
```

```python
def buscar_con_reranking(query, top_k_faiss=50, top_n_final=5):
    query_vector = model.encode([query]).astype('float32')
    distancias, indices = index.search(query_vector, top_k_faiss)
    
    candidatos = df_reducido.iloc[indices[0]].copy()
    
    # Reranking con Cross-Encoder
    pares = [[query, doc] for doc in candidatos['abstract_clean']]
    
    scores = reranker_model.predict(pares)
    
    candidatos['rerank_score'] = scores
    resultado_final = candidatos.sort_values(by='rerank_score', ascending=False)
    
    return resultado_final.head(top_n_final)
```

```python
query_usuario = "Advancements in large language models and transformers"
resultados_optimizados = buscar_con_reranking(query_usuario)

print(resultados_optimizados[['title', 'rerank_score']])
```

# 6. Evaluación del sistema

Ya que el corpus no contiene qrels, nos lo inventamos nosotros.

```python
qrels = {
    "large language models": ["123.456", "789.012"],
    "quantum computing": ["456.789"],
    "dark matter observations": ["111.222", "333.444"]
}
```

Definimos una función de cálculo de métricas para FAISS y Cross Encoder.

```python
def calcular_metricas(query, ids_relevantes, k=10):
    # Recuperación con FAISS
    query_vec = model.encode([query]).astype('float32')
    _, idx_faiss = index.search(query_vec, k)
    recuperados_faiss = df_reducido.iloc[idx_faiss[0]]['id'].tolist()
    
    # Recuperación con Reranker
    res_rerank = buscar_con_reranking(query, top_k_faiss=k, top_n_final=k)
    recuperados_rerank = res_rerank['id'].tolist()
    
    def evaluar(recuperados, relevantes):
        hits = [1 if r in relevantes else 0 for r in recuperados]
        p_at_k = sum(hits) / len(hits)
        
        mrr = 0
        for i, h in enumerate(hits):
            if h == 1:
                mrr = 1 / (i + 1)
                break
        return p_at_k, mrr

    met_faiss = evaluar(recuperados_faiss, ids_relevantes)
    met_rerank = evaluar(recuperados_rerank, ids_relevantes)
    
    return met_faiss, met_rerank
```

```python
# Ejecutar evaluación
for query, ids in qrels.items():
    f, r = calcular_metricas(query, ids)
    print(f"Query: {query}")
    print(f"  FAISS -> P@10: {f[0]:.2f}, MRR: {f[1]:.2f}")
    print(f"  Rerank -> P@10: {r[0]:.2f}, MRR: {r[1]:.2f}")
```
