# 1. Limpieza
Limpiar los documentos de caracteres que no corresponden.

```python
doc = df.iloc[0]["review"]
doc
```

Utilizamos un patrón para limpiar las etiquetas HTML.

```python
doc.replace("<br />", "").replace(">", "").replace("/", "")
```

Aquí tratamos de capturar las palabras que existen en el documento, pero aún hay algunas etiquetas que no se eliminaron por completo.

```python
doc.split()
```

Así que vamos a utilizar la librería `re` para usar una expresión regular que elimine todas las etiquetas HTML.

```python
import re
```

```python
re.sub(pattern=r"<.*?>", repl="", string=doc)
```

Si nos fijamos, las etiquetas se eliminaron correctamente, pero existen otros caracteres no deseados como puntos o comas. Vamos a complementar el método anterior para encontrar etiquetas HTML en una sola función.

```python
def clean_text(doc):
    # return re.sub(pattern=r"<.*?>", repl="", string=doc).replace(".", "").replace(",", "").replace("(", "").replace(")", "").replace("'", "").replace('"', '').replace("?", "").replace("[", "").replace("]", "").replace("!", "").replace("-", "").replace("/", " ").replace("@#$0&%", "").replace(":", "").replace(";", "")
    doc = re.sub(pattern=r"<.*?>", repl="", string=doc)
    return re.sub(pattern=r"[^a-zA-Z]+", repl=" ", string=doc) #Elimina caracteres no alfabéticos
```

```python
df["review"] = df["review"].apply(clean_text)
```

# 2. Normalización

Todo debe ser homogéneo, por lo tanto, vamos a convertir a minúsculas a todo el texto del corpus y obtener nuestro diccionario. Los caracteres que no nos interesan se eliminaron en el paso anterior.

```python
vocab = list(set(" ".join(df["review"].astype(str)).lower().split()))
```

```python
vocab[0:10]
```

```python
len(vocab)
```

# 3. Eliminación de StopWords

Ahora eliminaremos palabras vacías usando la librería nltk.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# nltk.download('stopwords')
# nltk.download('punkt')

stopwords = set(stopwords.words("english"))

clean_vocab = [word for word in vocab if word not in stopwords]
```

```python
clean_vocab[0:10]
```

```python
len(clean_vocab)
```

# 4. Stemming

Después de haber limpiado, debemos obtener las palabras que existen en el corpus y utilizaremos la técnica de stemming para devolver las palabras a una raíz. Esto reducirá nuestro diccionario de palabras.

```python
from nltk.stem import porter
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")
clean_vocab = set([stemmer.stem(word) for word in clean_vocab])
```

```python
len(clean_vocab)
```
