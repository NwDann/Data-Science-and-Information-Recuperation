### 1. Limpieza

Limpiar los documentos de caracteres que no corresponden

```python
doc = df.iloc[0]['review']
doc
```

```python
doc.replace('<', '').replace('>', '').replace('/', '')
```

```python
import re
```

```python
re.sub(pattern=r'<.*?>', repl='', string=doc)
```

```python
def clean_text(doc):
    return re.sub(pattern=r'<.*?>', repl='', string=doc).replace('.', ' ').replace(',','').replace('(', '').replace(')', '').replace('"', '').replace("'", '')
```

```python
df['review'].apply(clean_text)
```
